```python
import tensorflow as tf
from tensorflow.keras.callbacks import ModelCheckpoint, EarlyStopping
```

```python
checkpoint_callback = ModelCheckpoint(
    filepath='melhor_modelo.h5',        # nome do arquivo salvo
    monitor='val_accuracy',             # métrica a ser monitorada
    save_best_only=True,                # salva apenas o melhor
    mode='max',                         # "max" porque queremos a maior acurácia
    verbose=1                           # mostra mensagens quando salvar
)

early_stopping_callback = EarlyStopping(
    monitor='val_accuracy',  # Métrica a ser monitorada
    patience=10,             # Número de épocas sem melhoria antes de parar
    mode='max',              # "max" porque queremos a maior acurácia
    verbose=1,               # Exibe mensagens quando parar
    restore_best_weights=True  # Restaura os melhores pesos ao final
)
```

```python
model_dt = tf.keras.Sequential([
    tf.keras.layers.Input(shape=(X_train.shape[1],)),
    tf.keras.layers.Dense(32, activation='relu'),
    tf.keras.layers.Dense(16, activation='relu'),
    tf.keras.layers.Dense(3, activation='softmax')  # multiclasses
])

# Compilando o modelo
model_dt.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# Treinando o modelo e salvando o histórico
history = model_dt.fit(
    X_train, y_train,
    epochs=100,
    batch_size=32,
    validation_split=0.2,
    callbacks=[checkpoint_callback]
)
```

```python
# Gráficos de Acurácia e Loss
plt.figure(figsize=(10, 3.5))

# Acurácia
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Treinamento')
plt.plot(history.history['val_accuracy'], label='Validação')
plt.title('Acurácia por Época')
plt.xlabel('Época')
plt.ylabel('Acurácia')
plt.legend()

# Loss
plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Treinamento')
plt.plot(history.history['val_loss'], label='Validação')
plt.title('Erro (Loss) por Época')
plt.xlabel('Época')
plt.ylabel('Loss')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
# Predições do modelo
y_pred = model_dt.predict(X_test)

# Classificando as previsões para cálculo do classification report
y_pred_classes = np.argmax(y_pred, axis=1)  # Se estiver usando softmax

# Classification Report
print("Classification Report:")
print(classification_report(y_test, y_pred_classes, target_names=['Leve', 'Moderado', 'Vigoroso']))

# Matriz de Confusão
cm = confusion_matrix(y_test, y_pred_classes)

# Plotando a Matriz de Confusão
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=['Leve', 'Moderado', 'Vigoroso'], yticklabels=['Leve', 'Moderado', 'Vigoroso'])
plt.title('Matriz de Confusão')
plt.xlabel('Predições')
plt.ylabel('Verdades Reais')
plt.show()
```

CONVERSÃO PARA O TF LITE

```python
import os
```

```python
# Caminho do modelo Keras salvo
model_path = '/content/melhor_modelo.h5'
# Salvar o modelo Keras
model_dt.save(model_path)

# Converter o modelo Keras para TensorFlow Lite
tflite_converter = tf.lite.TFLiteConverter.from_keras_model(model_dt)
tflite_model = tflite_converter.convert()

# Salvar o modelo TensorFlow Lite em um arquivo
tflite_model_path = 'modelo_convertido.tflite'
with open(tflite_model_path, 'wb') as f:
    f.write(tflite_model)

# Comparar os tamanhos dos arquivos
keras_model_size = os.path.getsize(model_path) / (1024 * 1024)  # Tamanho em MB
tflite_model_size = os.path.getsize(tflite_model_path) / (1024 * 1024)  # Tamanho em MB

# Exibir os tamanhos
print(f"Tamanho do modelo Keras (em MB): {keras_model_size:.2f} MB")
print(f"Tamanho do modelo TensorFlow Lite (em MB): {tflite_model_size:.2f} MB")
```

```python
# prompt: Avalie a acurácia do modelo tflite. Para isso, teste a acurácia do modelo TensorFlow Lite, usando um código que utiliza tf.lite.Interpreter, carrega o modelo convertido e compara sua saída com a saída do modelo original em um conjunto de dados de teste.

import tensorflow as tf
import numpy as np

# Load the TFLite model and allocate tensors.
interpreter = tf.lite.Interpreter(model_path="modelo_convertido.tflite")
interpreter.allocate_tensors()

# Get input and output tensors.
input_details = interpreter.get_input_details()
output_details = interpreter.get_output_details()

# Test the model on random input data.
input_shape = input_details[0]['shape']
input_data = np.array(X_test, dtype=np.float32)

# Run predictions with the TFLite model
predictions_tflite = []
for i in range(len(input_data)):
    interpreter.set_tensor(input_details[0]['index'], np.expand_dims(input_data[i], axis=0))
    interpreter.invoke()
    output_data = interpreter.get_tensor(output_details[0]['index'])
    predictions_tflite.append(np.argmax(output_data))

predictions_tflite = np.array(predictions_tflite)


# Run predictions with the original Keras model
predictions_keras = np.argmax(model_dt.predict(X_test), axis=1)


# Compare the results
accuracy = np.mean(predictions_tflite == predictions_keras)
print(f"Acurácia do modelo TFLite em relação ao modelo Keras original: {accuracy * 100:.2f}%")

# Detailed comparison (optional)
mismatched_indices = np.where(predictions_tflite != predictions_keras)[0]
print(f"Indices onde as previsões não coincidem: {mismatched_indices}")

#Confusion Matrix
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

conf_matrix = confusion_matrix(y_test, predictions_tflite)
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=['Leve', 'Moderado', 'Vigoroso'], yticklabels=['Leve', 'Moderado', 'Vigoroso'])
plt.xlabel('Previstos')
plt.ylabel('Reais')
plt.title('Matriz de Confusão - TFLite')
plt.show()
```
